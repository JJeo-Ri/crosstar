### N = 3Ïù¥ÎùºÍ≥† Í∞ÄÏ†ïÌï† Í≤ΩÏö∞
- 1Ìöå Ï°∞Ìï©ÏúºÎ°ú ÎßåÎì§Ïñ¥ ÎÇº Ïàò ÏûàÎäî Ïàò : 3
- 2Ìöå Ï°∞Ìï©ÏúºÎ°ú ÎßåÎì§Ïñ¥ ÎÇº Ïàò ÏûàÎäî Ïàò : 6(3+3), 1(3/3), 9(3*3), 33(Ïó∞ÏÜç)


```python
def solution(N,number):
    possible_set = [0, [N]]
    if N == number : #
        return 1
    for i in range(2,9):                            #2~8ÍπåÏßÄÍ∞Ä Ï°∞Ìï©(9Î≤àÏù¥ÏÉÅ Ïì∞ÏßÄ ÏïäÏùå)
        case_set = []                               # Possible setÏóê appendÌï† caseset
        basic_num = int(str(N)*i)                   # NÏùò iÍ∞ú Ïó∞ÏÜçÌïú Ïà´Ïûê Ï∂îÍ∞Ä ex) 33, 333 , ...
        case_set.append(basic_num)
        for i_half in range(1,i//2+1):              #Ïà´ÏûêÎ•º Ï†àÎ∞òÍπåÏßÄÎßå ÏÇ¨Ïö©(Ï¢åÏö∞ÎåÄÏπ≠ÏúºÎ°ú Í∞ôÏùÄ Í≤∞Í≥ºÍ∞Ä ÎÇòÏò¥)
            for x in possible_set[i_half]:
                for y in possible_set[i-i_half]:    # x + y = i_half Ïù¥Î©∞ Î∞òÏúºÎ°úÎÇòÎà†ÏßÑ
                    print(possible_set)
                                                    # ÏÇ¨ÏπôÏó∞ÏÇ∞ÏùÑ ÏàòÌñâÌõÑ case_setÏóê ÎÑ£Ïùå
                    case_set.append(x+y)
                    case_set.append(x-y)
                    case_set.append(y-x)
                    case_set.append(x*y)
                    if y != 0:
                        case_set.append(x / y)
                    if x != 0:
                        case_set.append(y / x)
            if number in case_set:                   # case setÏóê ÏõêÌïòÎäî Ïà´ÏûêÍ∞Ä ÏûàÎã§Î©¥ Î∞òÌôò
                return i
            possible_set.append(case_set)
    return -1


print(solution(3,21))
```
